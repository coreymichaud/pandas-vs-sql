# Libraries

```python
# Insert python libraries
import pandas as pd
import duckdb
```

# Loading data into dataframe

```python
customers = pd.read_csv("customers.csv")
orders = pd.read_csv("orders.csv")
```

# Create SQL function

```python
# This makes it so the SQL cells look cleaner
def SQL(query: str):
    display(duckdb.sql(query).df())
```
